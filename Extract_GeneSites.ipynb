```python
# Utils

# function to reformat start coordinates
def reformatRefStart(object):
    arr = object.split(",")
    arr.remove("")
    arr = list(map(int, arr))
    arr = list(map(lambda x:x-2, arr))
    #print(arr)
    return arr

# function to reformat end coordinates
def reformatRefEnd(object):
    arr = object.split(",")
    arr.remove("")
    arr = list(map(int, arr))
    arr = list(map(lambda x:x+2, arr))
    #print(arr)
    return arr

def scan(chro,start):
    # Extract indices of chro
    chroIndices = sites.index[sites['Chr'] == str(chro)].tolist()
    # Go to corresponding index in mlist 
    for i in chroIndices:
        coords = mlist[i]
        #print(coords)
        # transform data in dataframe
        x = coords.to_frame()
        y = pd.DataFrame(columns = ['start', 'end'])
        y['start'] = x.index.get_level_values(0)
        y['end'] = x.index.get_level_values(1)
        #print(y)
        # extract if in range
        p = y[(start >= y['start']) & (start <= y['end'])]
        if not p.empty:
            return True
        
    return False
         
# process chunk 
def process(chunk):
    inRef = {"Index":[],"Chr":[],"Start":[], "Ref":[], "Alt":[], "Freq":[], "ID":[]}
    for row in chunk.itertuples():
        # get row info
        index = row[0]
        chro = row[1]
        start = row[2]
        ref = row[3]
        alt = row[4]
        freq = row[5]
        ID = row[6]  
        # scan in ref database
        result = scan(chro,start)  
        # if in ref database -> add to dictionary
        if  (result == True):
            print(result)
            inRef["Index"].append(index)
            inRef["Chr"].append(chro)
            inRef["Start"].append(start)
            inRef["Ref"].append(ref)
            inRef["Alt"].append(alt)
            inRef["Freq"].append(freq)
            inRef["ID"].append(ID)             
    df_inRef = pd.DataFrame.from_dict(inRef)
    return df_inRef
```

```python
# Load ref database
ref = pd.read_csv('/Volumes/Samsung_T5/hg19_refGene.csv', chunksize=1000)
ref = pd.concat(ref,ignore_index=True)
#print("ref shape:", ref.shape)
ref['Chr'] = ref['Chr'].str.replace('chr', '')
# Exon start/end coordinates
ExonStarts = ref["Exon starts"]
ExonEnds = ref["Exon ends"]
# reformat coordinates
reformatStart = lambda x: reformatRefStart(x)
ExonStarts = ExonStarts.apply(reformatStart)
reformatEnd = lambda x: reformatRefEnd(x)
ExonEnds = ExonEnds.apply(reformatEnd)
# Chromosome
refChr = ref["Chr"]
# Combine data
ExonStarts = pd.Series(ExonStarts)
refChr = pd.Series(refChr)
ExonEnds = pd.Series(ExonEnds)
sites = pd.concat([refChr, ExonStarts, ExonEnds], axis = 1)
sites = pd.DataFrame(sites)
#sites
```

```python
# Merge starts and ends from sites 

# Coord list
mlist = []

for index, row in sites.iterrows():
    rowStarts = np.array(sites.loc[index,"Exon starts"])
    rowEnds = np.array(sites.loc[index,"Exon ends"])
    #print(rowStarts)
    #print(rowEnds)
    array = [rowStarts,rowEnds]
    merge = pd.MultiIndex.from_arrays(array, names=('start', 'end'))
    #print(index)
    #print(merge)
    mlist.append(merge)
```
